```python
elements = []

while True:
    
    print("Elements in list")
    for element in elements:
        print("-> " + element)
        
    
    print("\nOptions")
    print("1. Add Elements")
    print("2. Delete Elements")
    print("3. Delete All Elements")
    print("4. Exit")
    print("Press any number e.g 1, 2, 3, 4")
    
    num = input("\nEnter any number ")
    
    if num == "1":
        elements.append(input("Enter your element: "))
    elif num == "2":
        elements.remove(input("Enter element which you want to delete"))
    elif num == "3":
        elements.clear()
    elif num == "4":
        break
    else:
        print("\nInvalid Input")
        
    print("\n%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%\n")
```
