```python
import requests
import bs4

#STEP:1
# GETTING THE HTML CONTENT

url = "https://www.daraz.pk"
file = requests.get(url)
html_content = file.content
```

```python
#html_content
```

```python
#STEP:2
# PARSING THE HTML CONTENT
# bs4 sets the content in the desired manner. (html syntax)

from bs4 import BeautifulSoup
soup = BeautifulSoup(html_content, "html.parser")
```

```python
#STEP:3
#HTML TREE TAVERSAL
#travelling the tree, firing queries, extracting and manipulating data.

#getting the title of html site
title = soup.title
```

```python
#getting all the anchor tags 
anchors = soup.find_all('a')
```

```python
#getting all the links from the anchors
#but we could not go any  where through these links because these are text

for links in anchors:
    print(links.get('href'))
```

```python

all_links = set()
for links in anchors:
    if(links.get('href') != '#'):
        link = "https://www.daraz.pk" +str(links.get('href'))
        all_links.add(link)
        print(link)
```
