```python
# (Edros-Renyi) ER Graph Model
# ER has two parameters (N,p)
# N is the number of nodes in the graph
# p is the probablity for any pair of node to be connected by any edge
# if the value of p is very small typical graph generated from the model tends to be sparse, meaning having few edges
# if the value of p is large typical graph generated from the model tends to be densely connected
```

```python
# Random graph model for a coin
# here p is single input argument , which is the probablity of success

from scipy.stats import bernoulli
bernoulli.rvs(p = 0.2 ) # rvs() generates single realization 
```

```python
N = 20
p = 0.2
#              PSUEDOCODE
# Create a empty graph
# Add all N nodes to graph
# Loop over all pair of nodes 
#     add an edge with probablity p

G = nx.Graph()
G.add_nodes_from(range(N))
for node1 in G.nodes():
    for node2 in G.nodes(): # As an edge is a pair of nodes
        if bernoulli.rvs(p=p):  # Here the first p is the name of the keyword argument and the 2nd p is the actual p 0.2
            G.add_edge(node1,node2)
```

```python
# As the graph is more dense because each node is repeating 
# So we set Node1 < Node2 
# And make it a function

N = 20
p = 0.2
#              PSUEDOCODE
# Create a empty graph
# Add all N nodes to graph
# Loop over all pair of nodes 
#     add an edge with probablity p
def er_graph(N,p):
    '''Generate an ER-Graph.'''
    G = nx.Graph()
    G.add_nodes_from(range(N))
    for node1 in G.nodes():
        for node2 in G.nodes(): # As an edge is a pair of nodes
            if node1 < node2 and bernoulli.rvs(p=p):  # Here the first p is the name of the keyword argument and the 2nd p is the actual p 0.2
                G.add_edge(node1,node2)
    return(G)
```

```python
nx.draw(er_graph(50,0.08), node_size = 40, node_color = 'gray' )
plt.savefig("er1.pdf")
```

```python
def plot_degree_distribution(G):
    plt.hist(list(G.degree().values()), histtype = "step" )
    plt.xlabel("degree $k$")
    plt.ylabel("$P(k)$")
    plt.tittle("Degree Distribution")
```

```python
G = er_graph(60,0.08)
plot_degree_distribution(G)
```

```python
# np.loadtxt is used to read adjacency matrices in the csv files
# adjacent matrix means the element ij is always same as ji, either both are zero or both are 1.

import numpy as np
A1 =  np.loadtxt("adj_allVillageRelationships_vilno_1.csv", delimiter = ",")
A2 =  np.loadtxt("adj_allVillageRelationships_vilno_2.csv", delimiter = ",")

# now converting the adjacency matrices to  graph objects
G1 = nx.to_networkx_graph(A1)
G2 = nx.to_networkx_graph(A2)
```

```python
# LCC(largest connecting component)
# max() for getting the maximum of the sequence
# max() take in a generator as its input

# Q:       But given two graph components, A and B, how can the max function possiblt know which is the maximum?
# Answer:  We need to tell the max function what number to associate with each object in sequence, in this case, a graph

# The size of the component is defined as the number of nodes it contains.
# We can obtain by applying the len function to a given component.

G1_LCC = max(nx.connected_component_subgraphs(G1), key = len)
G2_LCC = max(nx.connected_component_subgraphs(G2), key = len)
```

```python
print(len(G1_LCC))
print(G1_LCC.number_of_nodes())
```

```python
print(len(G2_LCC))
print(G2_LCC.number_of_nodes())
```

```python
# Computing the proportion of nodes that lie in the largest connected component for these two graphs

# About 97.9% of all the nodes of graph G1 are contained in the largest connected component.
print((G1_LCC.number_of_nodes() / G1.number_of_nodes())*100)

# About 92.3% of all the nodes of graph G2 are contained in the largest connected component.
print((G2_LCC.number_of_nodes() / G2.number_of_nodes())*100)
```
