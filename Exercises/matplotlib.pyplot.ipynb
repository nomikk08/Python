```python
import numpy as np
x = np.linspace(0,10,20)  # starts from o ends at 10 and has 20 points
y = x**2
print(x)
print(y)
```

```python
plt.plot([x,y]) # shows the graph of each element of (x,y)
plt.show()
```

```python
plt.plot(x,y)
plt.show()
```

```python
x = np.linspace(0,10,20)  # starts from o ends at 10 and has 20 points
y1 = x**2
y2 = x**1.5
plt.plot(x , y1, "bo-" , linewidth=3 , markersize=8) #bo-(blue circles)
```

```python
plt.plot(x , y2, "ys-" , linewidth=3 , markersize=8) #ys-(yellow squares)
```

```python
plt.plot([0,1,2],[0,1,4],"rd-") # rd-(red diamond)
```

```python
x = np.linspace(0,10,20) 
y1 = x**2
y2 = x**1.5
plt.plot(x , y1, "bo-" , linewidth=3 , markersize=8 , label = "first") 
plt.plot(x , y2, "rd-" , linewidth=3 , markersize=8 , label = "second")
plt.xlabel("X") # label the x axis and yaxis
plt.ylabel("Y")
plt.legend(loc = "upper left")# shows the legend on upper left corner
plt.savefig("myplot.pdf")
```

```python
x = np.logspace(-1,1,20) # log-1=1 , 1=10
y1 = x**2
y2 = x**1.5
plt.loglog(x , y1, "bo-" , linewidth=3 , markersize=8 , label = "first") 
plt.loglog(x , y2, "rd-" , linewidth=3 , markersize=8 , label = "second")
plt.xlabel("X") # label the x axis and yaxis
plt.ylabel("Y")
plt.legend(loc = "upper left")# shows the legend on upper left corner
plt.savefig("myplot.pdf")
```
