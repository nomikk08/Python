# SEABORN


RELATIONAL PLOTS   (relplot): scatterplot, lineplot
DISTRIBUTION PLOTS (deplot): histplot, kdeplot, ecdfplot, rugplot
CATEGORICAL PLOTS (catplot): stripplot, swarmplot, boxplot, violinplot, pointplot, barplot

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
import seaborn as sns

sns.set_theme() #load theme
tips = sns.load_dataset('tips') #loading a theme

sns.relplot( data =tips, x = 'total_bill', y = 'tip', col = 'time', hue ='smoker', style = 'smoker', size ='size')
```

```python
sns.displot(data = tips, x = 'total_bill', col = 'time', kde = True)
```

```python
sns.catplot(data = tips, kind = 'swarm', x = 'day', y = 'total_bill', hue = 'smoker' )
```

```python
sns.catplot(data = tips, kind = 'strip', x = 'day', y = 'total_bill', hue = 'smoker' )
```

```python
sns.catplot(data = tips, kind = 'box', x = 'day', y = 'total_bill', hue = 'smoker' )
```

```python
sns.catplot(data = tips, kind = 'point', x = 'day', y = 'total_bill', hue = 'smoker' )
```

```python
sns.catplot(data = tips, kind = 'violin', x = 'day', y = 'total_bill', hue = 'smoker' )
```

```python
sns.catplot(data = tips, kind = 'bar', x = 'day', y = 'total_bill', hue = 'smoker' )
```
